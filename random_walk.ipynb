### Em duas dimensões:

```python
nsteps = 5000

rngx = np.random.default_rng(seed=797845) #gerador de números aleatórios coordenada x
rngy = np.random.default_rng(seed=355972) #gerador de números aleatórios coordenada y

drawsx = rngx.integers(0, 2, size=nsteps)
drawsy = rngy.integers(0, 2, size=nsteps)


stepsx = np.where(drawsx == 0, -1, 1)
stepsy = np.where(drawsy == 0, -1, 1)

walkx = stepsx.cumsum()
walky = stepsy.cumsum()
```

```python
plt.plot(walkx, walky)
plt.show()
```

### Em três dimensões:

```python
from mpl_toolkits.mplot3d import Axes3D

nsteps = 5000

rngx = np.random.default_rng(seed=797845) #gerador de números aleatórios coordenada x
rngy = np.random.default_rng(seed=355972) #gerador de números aleatórios coordenada y
rngz = np.random.default_rng(seed=569835) #gerador de números aleatórios coordenada z

drawsx = rngx.integers(0, 2, size=nsteps)
drawsy = rngy.integers(0, 2, size=nsteps)
drawsz = rngz.integers(0, 2, size=nsteps)


stepsx = np.where(drawsx == 0, -1, 1)
stepsy = np.where(drawsy == 0, -1, 1)
stepsz = np.where(drawsz == 0, -1, 1)

walkx = stepsx.cumsum()
walky = stepsy.cumsum()
walkz = stepsz.cumsum()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(walkx, walky, walkz)
plt.show()
```

### Em uma dimensão (duas moedas):

```python
nsteps = 5000

rng1 = np.random.default_rng(seed=797845) #gerador de números aleatórios da partícula 1
rng2 = np.random.default_rng(seed=897643) #gerador de números aleatórios da partícula 2


draws1 = rng1.integers(0, 2, size=nsteps)
draws2 = rng2.integers(0, 2, size=nsteps)


steps1 = np.where(draws1 == 0, 1, -1)
steps2 = np.where(draws2 == 0, 1, -1)

walk10 = 0 #define a posição inicial da partícula 1
walk20 = 5 #define a posição inicial da partícula 2
walk1 = steps1.cumsum()
walk2 = steps2.cumsum()
```

```python
plt.plot(walk1)
plt.plot(walk2) 
plt.show()
```

```python
walk10 = 0
walk20 = 4

x1 = walk10 + walk1
x2 = walk20 + walk2
same_point = np.argwhere(x1 == x2) #quando as partículas estiverem no mesmo ponto

print(same_point[0]) #primeiro ponto de mesma posição
```

```python
print(same_point[-1]) #último ponto de mesma posição
```

```python
len(same_point) #quantas vezes as partículas estiveram no mesmo ponto
```

```python
plt.plot(x1[:260])
plt.plot(x2[:260]) 
plt.show()
```

### Em duas dimensões (duas moedas):

```python
nsteps = 5000

rngx1 = np.random.default_rng(seed=797845) #gerador de números aleatórios coordenada x da moeda 1
rngy1 = np.random.default_rng(seed=355972) #gerador de números aleatórios coordenada y da moeda 1

drawsx1 = rngx1.integers(0, 2, size=nsteps)
drawsy1 = rngy1.integers(0, 2, size=nsteps)


stepsx1 = np.where(drawsx1 == 0, -1, 1)
stepsy1 = np.where(drawsy1 == 0, -1, 1)

walkx1 = stepsx1.cumsum()
walky1 = stepsy1.cumsum()


rngx2 = np.random.default_rng(seed=797631674) #gerador de números aleatórios coordenada x da moeda 2
rngy2 = np.random.default_rng(seed=197987421) #gerador de números aleatórios coordenada y da moeda 2

drawsx2 = rngx2.integers(0, 2, size=nsteps)
drawsy2 = rngy2.integers(0, 2, size=nsteps)


stepsx2 = np.where(drawsx2 == 0, -1, 1)
stepsy2 = np.where(drawsy2 == 0, -1, 1)

walkx2 = stepsx2.cumsum()
walky2 = stepsy2.cumsum()
```

```python
plt.plot(walkx1, walky1)
plt.plot(walkx2, walky2)
plt.show()
```
