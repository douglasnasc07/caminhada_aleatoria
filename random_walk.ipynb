### Caminhada aleatória:

### Em uma dimensão:

```python
import numpy as np
import matplotlib.pyplot as plt

total_passos = 1000 #número de passos

gerador = np.random.default_rng(seed=2) #gerador de números aleatórios

#gera uma matriz de total_passos números inteiros aleatórios, cada um 0 ou 1:
zero_um = gerador.integers(0, 2, size=total_passos)

#np.where gera um novo array passos:
#Se o elemento for 0, o elemento correspondente stepsserá definido como 1:
#Se o elemento for 1, o elemento correspondente stepsserá definido como -1:
passos = np.where(zero_um == 0, 1, -1)

x0 = 2 #define a posição inicial como 3 = (4-1)

x = x0 + passos.cumsum()

```

```python
ponto_zero = np.argwhere(x == 0)
print(ponto_zero) #número de passos até passar por zero
```

```python
print(len(ponto_zero)) #número de vezes que passa por zero
```

```python
print(ponto_zero[0]) #número de passos para passar pela primeira vez por zero
```

```python
print(ponto_zero[-1]) #número de passos para passar pela última vez por zero
```

```python
x.min()
```

```python
x.max()
```

```python
x[-1] - x0 #distância final em passos em relação ao ponto de partida
```

```python
x.max() - x.min() #amplitude
```

```python
plt.figure(figsize=(20,10))
plt.plot(x)
plt.show()
```

### Em duas dimensões:

```python
total_passos = 1000

gerador_x = np.random.default_rng(seed=25) #gerador de números aleatórios coordenada x
gerador_y = np.random.default_rng(seed=90) #gerador de números aleatórios coordenada y

zero_um_x = gerador_x.integers(0, 2, size=total_passos)
zero_um_y = gerador_y.integers(0, 2, size=total_passos)

passos_x = np.where(zero_um_x == 0, -1, 1)
passos_y = np.where(zero_um_y == 0, -1, 1)

x0 = 4 #define a posição inicial como 3 = (4-1)
y0 = 3

x = x0 + passos_x.cumsum()
y = y0 + passos_y.cumsum()
```

```python
plt.figure(figsize=(20,10))
plt.plot(x, y)



plt.show()
```

```python
x[-1]
```

```python
R = np.sqrt((x-x0)**2 + (y-y0)**2) #distância em relação ao ponto de partida em passos
print(R)
```

```python
ponto_zero = np.argwhere(R == 0)
print(ponto_zero)
```

```python
print(len(ponto_zero)) #número de vezes que passa pelo ponto inicial
```

```python
#criando uma reta que liga o ponto inicial ao ponto final

x1, y1 = x0, y0  # Coordenadas do primeiro ponto
x2, y2 = x[-1], y[-1]  # Coordenadas do segundo ponto
t = np.linspace(x1, x2, 100)
m = (y2 - y1) / (x2 - x1)  # Cálculo da inclinação (coeficiente angular)
b = y1 - m * x1  # Cálculo do coeficiente linear
s = m * t + b  # Equação da reta
```

```python
plt.figure(figsize=(20,10))
plt.plot(x, y)
plt.plot(t, s)
plt.plot(x1, y1, 'ro', label='Ponto inicial')  # Marca o primeiro ponto em vermelho
plt.plot(x2, y2, 'go', label='Ponto final')  # Marca o segundo ponto em azul
plt.legend()
plt.show()
```

```python
np.sqrt((x[-1]-x0)**2 + (y[-1]-y0)**2 ) #distância do ponto inicial ao ponto final em passos
```

### Em três dimensões:

```python
from mpl_toolkits.mplot3d import Axes3D

nsteps = 1000

gerador_x = np.random.default_rng(seed=25) #gerador de números aleatórios coordenada x
gerador_y = np.random.default_rng(seed=90) #gerador de números aleatórios coordenada y
gerador_z = np.random.default_rng(seed=70) #gerador de números aleatórios coordenada z


zero_um_x = gerador_x.integers(0, 2, size=total_passos)
zero_um_y = gerador_y.integers(0, 2, size=total_passos)
zero_um_z = gerador_z.integers(0, 2, size=total_passos)

passos_x = np.where(zero_um_x == 0, -1, 1)
passos_y = np.where(zero_um_y == 0, -1, 1)
passos_z = np.where(zero_um_z == 0, -1, 1)

x0 = 4 #define a posição inicial como 3 = (4-1)
y0 = 3
z0 = 1

x = x0 + passos_x.cumsum()
y = y0 + passos_y.cumsum()
z = z0 + passos_z.cumsum()
```

```python

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z)
plt.plot(x0, y0, z0, 'ro', label='Ponto inicial') 
plt.plot(x[-1], y[-1], z[-1], 'go', label='Ponto final') 
plt.legend()
plt.show()
```

```python
np.sqrt((x[-1]-x0)**2 + (y[-1]-y0)**2 + (z[-1]-z0)**2 ) #distância do ponto inicial ao ponto final em passos
```

```python
d = np.sqrt((x-x0)**2 + (y-y0)**2 + (z-z0)**2 )
ponto_zero = np.argwhere(d == 0)
print(ponto_zero)
```

```python
print(len(ponto_zero)) #número de vezes que passa pelo ponto inicial
```

### Em uma dimensão (duas moedas):

```python
total_passos = 1000

gerador_1 = np.random.default_rng(seed=25) #gerador de números aleatórios para moeda 1
gerador_2 = np.random.default_rng(seed=90) #gerador de números aleatórios para moeda 2

zero_um_1 = gerador_1.integers(0, 2, size=total_passos)
zero_um_2 = gerador_2.integers(0, 2, size=total_passos)

passos_1 = np.where(zero_um_1 == 0, -1, 1)
passos_2 = np.where(zero_um_2 == 0, -1, 1)

x01 = 4 #define a posição inicial como 3 = (4-1)
x02 = 0

x1 = x01 + passos_1.cumsum()
x2 = x02 + passos_2.cumsum()
```

```python
plt.figure(figsize=(20,5))
plt.plot(x1, label='moeda 1')
plt.plot(x2,label='moeda 2')
plt.legend()
plt.show()
```

```python
mesmo_ponto = np.argwhere(x1==x2)

print(len(mesmo_ponto)) #quantas vezes se encontram
```

```python
print(mesmo_ponto[0]) #quando se encontram pela primeira vez
```

```python
print(mesmo_ponto[-1]) #quando se encontram pela última vez
```

```python
x01 - x02 #distância inicial em passos
```

```python
x1[-1] - x2[-1] #distância final em passos
```

```python
encounter_idx = np.argwhere(x1==x2).ravel()
steps = range(len(x1))
```

```python
plt.figure(figsize=(20,5))
plt.plot(steps,x1,label='moeda 1')
plt.plot(steps,x2,label='moeda 2')
for i, idx in enumerate(encounter_idx):
    if i==0:
        label = 'encontro'
    else:
        label = None
    plt.vlines(
        steps[idx],
        ymin=min(x1.min(),x2.min()),
        ymax=max(x1.max(),x2.max()),
        linestyle='dashed',
        linewidth=1,
        color='k',
        label=label
    )

plt.ylabel("Posição")
plt.xlabel("Passos")
plt.xlim(-1,1000)
plt.legend()
plt.show
```

### Em duas dimensões (duas moedas):

```python
total_passos = 1000

gerador_1x = np.random.default_rng(seed=25)
gerador_1y = np.random.default_rng(seed=30)

gerador_2x = np.random.default_rng(seed=35)
gerador_2y = np.random.default_rng(seed=40)

zero_um_1x = gerador_1x.integers(0, 2, size=total_passos)
zero_um_1y = gerador_1y.integers(0, 2, size=total_passos)

zero_um_2x = gerador_2x.integers(0, 2, size=total_passos)
zero_um_2y = gerador_2y.integers(0, 2, size=total_passos)

passos_1x = np.where(zero_um_1x == 0, -1, 1)
passos_1y = np.where(zero_um_1y == 0, -1, 1)

passos_2x = np.where(zero_um_2x == 0, -1, 1)
passos_2y = np.where(zero_um_2y == 0, -1, 1)

x01 = 0
y01 = 0

x02 = 1
y02 = 1


x1 = x01 + passos_1x.cumsum()
y1 = y01 + passos_1y.cumsum()

x2 = x02 + passos_2x.cumsum()
y2 = y02 + passos_2y.cumsum()
```

```python
plt.figure(figsize=(20,10))
plt.plot(x1, y1, label='moeda 1')
plt.plot(x2, y2, label='moeda 2')
plt.plot(x01, y01, 'bo', label='Ponto inicial moeda 1')
plt.plot(x1[-1], y1[-1], 'ro', label='Ponto final moeda 1') 
plt.plot(x02, y02, 'go', label='Ponto inicial moeda 2')
plt.plot(x2[-1], y2[-1], 'mo', label='Ponto final moeda 2') 
plt.legend()

plt.show()
```

```python
d = np.sqrt((x2 - x1)**2 + (y2-y1)**2)
mesmo_ponto = np.argwhere(d==0)
```

```python
print(len(mesmo_ponto)) #número de vezes que se encontram
```

```python
np.sqrt((x2[0] - x1[0])**2 + (y2[0]-y1[0])**2) #distância inicial entre as moedas
```

```python
np.sqrt((x2[-1] - x1[-1])**2 + (y2[-1]-y1[-1])**2) #distância final entre as moedas
```

```python
gerador_1x = np.random.default_rng(seed=25)
gerador_1y = np.random.default_rng(seed=30)
gerador_1z = np.random.default_rng(seed=50)

gerador_2x = np.random.default_rng(seed=35)
gerador_2y = np.random.default_rng(seed=40)
gerador_2z = np.random.default_rng(seed=55)


zero_um_1x = gerador_1x.integers(0, 2, size=total_passos)
zero_um_1y = gerador_1y.integers(0, 2, size=total_passos)
zero_um_1z = gerador_1z.integers(0, 2, size=total_passos)


zero_um_2x = gerador_2x.integers(0, 2, size=total_passos)
zero_um_2y = gerador_2y.integers(0, 2, size=total_passos)
zero_um_2z = gerador_2z.integers(0, 2, size=total_passos)


passos_1x = np.where(zero_um_1x == 0, -1, 1)
passos_1y = np.where(zero_um_1y == 0, -1, 1)
passos_1z = np.where(zero_um_1z == 0, -1, 1)


passos_2x = np.where(zero_um_2x == 0, -1, 1)
passos_2y = np.where(zero_um_2y == 0, -1, 1)
passos_2z = np.where(zero_um_2y == 0, -1, 1)


x01 = 0
y01 = 0
z01 = 0

x02 = 1
y02 = 1
z02 = 1


x1 = x01 + passos_1x.cumsum()
y1 = y01 + passos_1y.cumsum()
z1 = z01 + passos_1z.cumsum()

x2 = x02 + passos_2x.cumsum()
y2 = y02 + passos_2y.cumsum()
z2 = z02 + passos_2z.cumsum()
```
